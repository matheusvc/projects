```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
nome_arquivo = str(input('Digite o nome do arquivo baixado do semrush: '))

```

```python
semrush = pd.read_csv(nome_arquivo)
semrush
```

```python
semrush_filtro = semrush[['Keyword','Position', 'Search Volume', 'Url']]


                      
```

```python
path = str(input('Digite a Url: '))
```

```python
semrush_path = semrush_filtro[semrush_filtro['Url'].str.match(path)]
semrush_path.head()
```

```python
x = ['Top 100', 'Top 50', 'Top 20', 'Top 10', 'Top 3'] 
```

```python
def top(x):
    if x <=3:
        return 'Top 3'
    elif x<=10:
        return 'Top 10'
    elif x<=20:
        return 'Top 20'
    elif x<=50:
        return 'Top 50'
    else:
        return 'Top 100'
```

```python
semrush_path['Top'] = semrush_path['Position'].apply(top)
```

```python
semrush_path.head()
```

```python
fig, ax = plt.subplots()
ax.bar(x,semrush_path['Top'].value_counts())
```
