```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
backlink = pd.read_csv("C:/Users/SAMSUNG/Downloads/2018-04-19_14-48_explorer-backlinks_fresh_domain_tubonasa-com-br.csv")
```

```python
#### Anchor Text ######
backlink['Quantidade'] = 1
backlink['Porcentagem AnchorText (%)'] = (backlink['Quantidade'] /  backlink['Quantidade'].sum())*100
texto_ancora = backlink[['AnchorText', 'Quantidade', 'Porcentagem AnchorText (%)']]
y = pd.pivot_table(
    texto_ancora, index = ['AnchorText'], values = ['Quantidade', 'Porcentagem AnchorText (%)'], aggfunc = {'Quantidade':np.sum, 'Porcentagem AnchorText (%)':np.sum}).sort_values(by = ['Quantidade'],ascending=False)
```

```python
plt.pie(backlink['AnchorText'].value_counts(), autopct='%1.1f%%' )
plt.figure(figsize = (50, 50))
plt.show()
```
